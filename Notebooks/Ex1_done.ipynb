# Simple plot

Import the necessary libraries to deal with vectorized arrays and to plot data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Be sure that the figure that will be plotted in this Notebook will appear in it.

```python
%matplotlib inline
```

The following cell will write a file on your disk.

```python
%%writefile data_ex1.dat
# This file is used in a Notebook on numpy and matplotlib
# N  x  y  I
1  3  5  8.0
2  5  7  10.0
3  7  2  6.0
4  9  -1 3.0
5  11 1 12.0
```

Read the file previously writen in the variable d, which is a numpy structured array, with the names corresponding to the values in the 2nd line of the data file.

```python
d = np.genfromtxt('data_ex1.dat', names=True, comments='#', dtype=None, skip_header=1)
```

Print the values of d, check that the type for the 3 first fields are interger and for the last one is float.

```python
print d
print d.dtype
```

Plot the values from d in a diagram where "x" and "y" are used as coordinates and "I" as the color of the symbol. Make a nice plot where the symbols have a decent size. Don't forget to put a colorbar.

```python
f, ax = plt.subplots()
sc = ax.scatter(d['x'], d['y'], c=d['I'], s=100, edgecolor="none")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Exmple 1')
cb = f.colorbar(sc)
cb.set_label('I');
```

What do you think about the color table in this case? Change the plot to make a better one.

```python
# It's a little hard to visualize the relation between the colors and the value when there is so few points.
# Better use a logically increasing color table, "winter" for example
f, ax = plt.subplots()
sc = ax.scatter(d['x'], d['y'], c=d['I'], s=100, edgecolor="none", cmap="winter")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Example 2')
cb = f.colorbar(sc)
cb.set_label('I');
```

Plot a BPT diagram from these data. On the x axis, put NII6584/Halpha and on the y axis put OIII5007/Hbeta.

```python
f, ax = plt.subplots()
ax.scatter(d['N__2__6584A']/d['H__1__6563A'], d['O__3__5007A']/d['H__1__4861A'] )
ax.set_xlabel(r'[NII]/H$\alpha$')
ax.set_ylabel(r'[OIII]/H$\beta$')
ax.set_title('Example 3');
```

Compare yout results with this one: http://www.aanda.org/articles/aa/full_html/2010/11/aa13886-09/img26.png

Change your code to reproduce the same BPT plot.

```python
f, ax = plt.subplots()
x = d['N__2__6584A']/d['H__1__6563A']
y = d['O__3__5007A']/d['H__1__4861A']
ax.scatter(np.log10(x), np.log10(y))
ax.set_xlabel(r'log [NII]/H$\alpha$')
ax.set_ylabel(r'log [OIII]/H$\beta$')
ax.set_title('Example 4');
```

make a mask to select only the points where log(OIII/Hb) > -1.25

```python
mask = np.log10(y) > -1.25
```

plot the selected data, using the field "OXYGEN" for the color code.

```python
f, ax = plt.subplots()
x = (d['N__2__6584A']/d['H__1__6563A'])[mask]
y = (d['O__3__5007A']/d['H__1__4861A'])[mask]
c = d['OXYGEN'][mask]
sc = ax.scatter(np.log10(x), np.log10(y), c=c, edgecolor='none')
cb = f.colorbar(sc)
ax.set_xlabel(r'log [NII]/H$\alpha$')
ax.set_ylabel(r'log [OIII]/H$\beta$')
ax.set_title('Example 5')
cb.set_label('O/H');
```

Plot the selected data, using the field "logU_mean" for the color code. Save the resulting figure in PDF and png format.

```python
f, ax = plt.subplots()
x = (d['N__2__6584A']/d['H__1__6563A'])[mask]
y = (d['O__3__5007A']/d['H__1__4861A'])[mask]
c = d['logU_mean'][mask]
sc = ax.scatter(np.log10(x), np.log10(y), c=c, edgecolor='none')
cb = f.colorbar(sc)
ax.set_xlabel(r'log [NII]/H$\alpha$')
ax.set_ylabel(r'log [OIII]/H$\beta$')
ax.set_title('Example 5')
cb.set_label('log(U)')
f.savefig('BPT1.pdf')
f.savefig('BPT1.png')
```

Define a mask where the field "pop1" is > 0.5. Plot the histogram of NII6584/Ha for the data corresponding to the mask, and for the complementary data.

```python
mask_pop = d['pop1'] > 0.5
f, ax = plt.subplots()
x = np.log10(d['N__2__6584A']/d['H__1__6563A'])
ax.hist((x[mask_pop], x[~mask_pop]), bins=50, histtype='step')
ax.set_xlabel(r'[NII]/H$\alpha$')
ax.set_ylabel('N');
```
