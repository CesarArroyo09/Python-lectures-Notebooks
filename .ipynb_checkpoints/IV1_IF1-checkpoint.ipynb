```python
%matplotlib inline
import pyneb as pn
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I need CV1. I will obtain it by using C_V1/C_4959 = 6.56e-5 (PP13) and PyNeb 
# Define an O++ atom to obtain the emissivity of 4959
o3 = pn.Atom('O',3)
# define an array of temperature from 1e3 to 2e4
t = np.logspace(3, 4.2, 100)
# Plot the emissivity of 4959 from PyNeb and from the PP13 formula, 
# looking for the C_4959 that make them equals.
plt.semilogy(t, o3.getEmissivity(t, 1e2, wave=4959))
# Here I look for C4959 using trial/error and formulae of PP13
def IF1_test(t, C): 
    return C * (t/1e4)**-0.34*np.exp(-29160/t) 
plt.semilogy(t, IF1_test(t, 2.1e-20)) # finaly 2.1e-20 is found. Check that this overplot with PyNeb. 
plt.xlabel('T[K]')
plt.ylabel('[OIII]4959 emissivity')
```

```python
# Defining the I_V1(T) function:
def IV1(t):
    # from PP13, and using CV1 = 2.1e-20 * 6.56e-5
    return 1.37e-24 * (t/1e4)**-0.755
def IF1(t): 
    # Using the C determined above
    return 2.1e-20 * (t/1e4)**-0.34 * np.exp(-29160/t) 
# Compare my value to eq. 1 value:
coeff = 1.37e-24/2.1e-20/4/1.772e-5 # Something strange here, it should be 1.00
print coeff
```

```python
# Defining a function that will return the observed ratio I(V1)/I/F1):
def rapI(T1, T2, w1, w2):
    # This will return the observed ratio I(V1)/I/F1) in case of 
    # 2 regions of different Te and weighting w1, w2 
    # w is something like ne * nO+ * Volume
    return 1e5*(IV1(T1)*w1 + IV1(T2)*w2) / (IF1(T1)*w1 + IF1(T2)*w2) / coeff / 4.
```

```python
# Print I(V1)/I(F1) in case of 2 Te:
print rapI(8000, 10000, 1, 1) # 2 regions of Te 8000 and 10000, equal weigth:
```

```python
# 2 regions of Te 2000 and 10000, 
# the hotest weigthing 17 times more than the coolest.
# This leads to the same result as the previous...
print rapI(500, 10000, 1, 17.0) 
```

```python
# Define a function with more parameters to play with
def rapI_bis(T1, T2, n1, n2, v1, v2, abo1, abo2):
    # This will return the observed ratio I(V1)/I/F1) in case of 2 regions 
    # of different Te, n, v and abo
    # being electron temperature, electron density, volume, and O/H resp.
    # w is then something like ne**2 * O/H * Volume
    w1 = n1**2 * v1 * abo1
    w2 = n2**2 * v2 * abo2
    return 1e5*(IV1(T1)*w1 + IV1(T2)*w2) / (IF1(T1)*w1 + IF1(T2)*w2) / coeff / 4.
```

```python
# small temperature contraste, small density contraste, same volume, same abundances
print rapI_bis(T1=10000, T2=12000, n1=1e4, n2=3e3, 
               v1=1, v2=1., abo1=10**-3.3, abo2=10**-3.3)
```

```python
# Large temperature contraste, same density contraste, 
# huge volume contraste, large abundance contraste
# The same ratio!!!
print rapI_bis(T1=1000, T2=12000, n1=1e4, n2=3e3, 
               v1=0.0001, v2=1., abo1=10**-2.3, abo2=10**-3.3)
```
